```python
import numpy as np
import matplotlib.pyplot as plt

def plot_surface(data, title):
    num_points = data.shape[0]
    
    # Create X and Y coordinate arrays
    x = np.linspace(-1, 1, num_points)
    y = np.linspace(-1, 1, num_points)
    X, Y = np.meshgrid(x, y)

    # Define circular mask (radius 1)
    R = np.sqrt(X**2 + Y**2)
    mask = R <= 1

    # Mask out points outside the circle
    data_masked = np.where(mask, data, np.nan)

    # Create the 3D figure
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    plt.title(title)

    print(data)
    # Plot the surface
    surf = ax.plot_surface(X, Y, data_masked, cmap='viridis')

    # Add a color bar for reference
    # fig.colorbar(surf)

    # Show the plot
    plt.show()
```

## Plot arbitrary surface given coefficients

```python
def plot_surface_given_coeffs(zernike_coeffs):
    # ugly, but works
    modes = [z_0_0, z_1_1m, z_1_1, z_2_2m, z_2_0, z_2_2, z_3_3m, z_3_1m, z_3_1, z_3_3]
    surface = np.zeros_like(z_0_0)
    for i in range(len(modes)):
        surface += zernike_coeffs[i] * modes[i]
    
    print(f'Coefficients: {zernike_coeffs}')
    plot_surface(surface, 'Surface')

def normalize_coefficients(coeffs):
    return coeffs / np.sum(coeffs)
```

## Create smoothly changing wavefronts

```python
import numpy as np

def numpy_zoom(arr, zoom_factor):
    """
    Approximate scipy.ndimage.zoom using only NumPy.
    Performs bilinear interpolation for 2D arrays.
    """
    if arr.ndim != 2:
        raise ValueError("This function currently supports only 2D arrays.")

    # Original coordinates
    h, w = arr.shape
    y = np.arange(h)
    x = np.arange(w)

    # Target size
    new_h = int(np.round(h * zoom_factor))
    new_w = int(np.round(w * zoom_factor))

    # New coordinates (normalized to original index space)
    y_new = np.linspace(0, h - 1, new_h)
    x_new = np.linspace(0, w - 1, new_w)
    x_new_grid, y_new_grid = np.meshgrid(x_new, y_new)

    # Floor and ceil coordinates
    x0 = np.floor(x_new_grid).astype(int)
    x1 = np.clip(x0 + 1, 0, w - 1)
    y0 = np.floor(y_new_grid).astype(int)
    y1 = np.clip(y0 + 1, 0, h - 1)

    # Distances
    dx = x_new_grid - x0
    dy = y_new_grid - y0

    # Bilinear interpolation
    top = (1 - dx) * arr[y0, x0] + dx * arr[y0, x1]
    bottom = (1 - dx) * arr[y1, x0] + dx * arr[y1, x1]
    result = (1 - dy) * top + dy * bottom

    return result

# Example usage
small = np.array([
    [1, 8, 4],
    [4, 5, 2],
    [7, 2, 6]
])

large = numpy_zoom(small, 10)

# Optional: visualize
plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(small, cmap='viridis', origin='lower')
plt.subplot(1, 2, 2)
plt.title("Extrapolated")
plt.imshow(large, cmap='viridis', origin='lower')
plt.show()

```

```python
wavefront_seed = np.random.randint(-1, 2, (3,3))
wavefront = numpy_zoom(wavefront_seed, 10)

plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plt.title("Seed")
plt.imshow(wavefront_seed, cmap='viridis', origin='lower')
plt.subplot(1, 2, 2)
plt.title("Wavefront")
plt.imshow(wavefront, cmap='viridis', origin='lower')
plt.show()
```

```python
from IPython.display import clear_output

new_wavefront = wavefront
fig, ax = plt.subplots(1, 1)
im = ax.imshow(new_wavefront, cmap='viridis', origin='lower')

timesteps = 10
for i in range(timesteps):
    noise_seed = (np.random.random((3,3)) - 0.5) * 2
    noise = numpy_zoom(noise_seed, 10)
    new_wavefront += noise

    im = plt.imshow(new_wavefront, cmap='viridis', origin='lower')
    plt.colorbar(im)
    plt.pause(0.1)
    clear_output(wait=True)
```
